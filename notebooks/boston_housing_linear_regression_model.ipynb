```python
#Import Libraries
import pandas as pd
import numpy as np
#import xgboost as xgb
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
#import shap
#shap.initjs()
```

```python
#Constants
data_directory='../data/'
```

```python
#Create dataframe for the ground truth data set
df_gt_boston=pd.read_csv(data_directory+'Boston.csv')
df_gt_boston
```

```python
#Split train and test data set sizes

train_size=70
test_size=100-train_size

train_data=df_gt_boston.sample(frac=train_size/100,random_state=200) #random state is a seed value
test_data=df_gt_boston.drop(train_data.index)
```

```python
#Train Data
train_data.head(3)
```

```python
#Test Data
test_data.head(3)
```

```python
# Training dataset 
#features
x_train=train_data.loc[:, 'crim':'lstat']
#predictor
y_train=train_data['medv']

# Test dataset 
#features
x_test=test_data.loc[:, 'crim':'lstat']
#predictor
y_test=test_data['medv']



num_train_val=len(x_train)
num_test_val=len(x_test)
print("Train Values : %d" % (num_train_val))
print(" Test Values : %d" % (num_test_val))
```

```python
# Instantiate Linear Regression model
housing_predictor = LinearRegression()
```

```python
# Train the linear regression predictor model to fit the training data
housing_predictor.fit(x_train,y_train)
```

```python
# Test the trained model
preds_LR = housing_predictor.predict(x_test)
predictions = [round(value,2) for value in preds_LR]
print ('Example Predicted value (vs) Actual Value')
print("Predicted Value : %f" % (predictions[0]) )
print("   Actual Value : %f" % (y_test[0]))
```

```python
# Measure Accuracy of Prediction

rmse = np.sqrt(mean_squared_error(y_test, predictions))
rounded_scaled = round(rmse*1000,0)
range_val=round((max(y_test)-min(y_test))*1000,0)

print("         Root Mean Square Error (RMSE) :$%f" % (rounded_scaled) )
print("Range of Actual Values in Test dataset :$%f" % (range_val))
print("")
print("Model is able to predict the price of houses within the amount of :$%f" %(rounded_scaled))
```
